# **Training Data**

Training data adalah kumpulan data yang digunakan untuk melatih model machine learning. Data latihan ini membantu model machine learning untuk belajar bagaimana melakukan tugas yang diinginkan secara otomatis. Training data sangat penting dalam mengajarkan algoritma atau model machine learning untuk memprediksi hasil yang diinginkan. Data latihan ini juga dapat diperkaya dengan pelabelan atau anotasi data jika menggunakan supervised learning atau pendekatan hibrida yang melibatkan supervised learning. Selain itu, terdapat pula test data yang digunakan untuk mengukur kinerja algoritma yang digunakan untuk melatih mesin. Kedua jenis data, baik training maupun test data, sangat penting untuk meningkatkan dan memvalidasi model machine learning.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
```

```python
file_path = '/content/Sleep_health_and_lifestyle_dataset.csv'
df = pd.read_csv(file_path)

# Pilih fitur-fitur yang akan digunakan untuk prediksi
features = df[['Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'Heart Rate']]  # Ganti 'Fitur1', 'Fitur2', ... dengan nama fitur sesuai dataset Anda

# Pilih kolom target yang ingin diprediksi (misalnya, 'Gender')
target = df['Heart Rate']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Inisialisasi model (misalnya, menggunakan Random Forest Classifier)
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Latih model pada data latih
model.fit(X_train, y_train)

# Lakukan prediksi pada data uji
predictions = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f'Accuracy: {accuracy}')
print('Classification Report:\n', report)
```

```python
# Import pustaka yang diperlukan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')

# Menentukan variabel independen (X) dan variabel dependen (y)
X = data.drop('Quality of Sleep', axis=1)  # Ganti 'target_variable' dengan nama variabel dependen
y = data['Heart Rate']

# Melakukan Label Encoding pada fitur kategori
label_encoder = LabelEncoder()
X_encoded = X.apply(label_encoder.fit_transform)

# Memisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Membuat model machine learning (gantilah dengan model yang sesuai)
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Melakukan prediksi pada data uji
y_pred = model.predict(X_test)

# Evaluasi performa model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

# Menampilkan laporan klasifikasi
print('Classification Report:')
print(classification_report(y_test, y_pred))
```
