```python
import pandas as pd #printah untuk mengambil library pandas

# Tentukan path ke file CSV atau dataset lainnya di Google Drive Anda
file_path = "/content/drive/MyDrive/dataset/Social_Network_Ads.csv" #pemanggilan dataset yang berada pada drive

# Baca dataset menggunakan pandas
df = pd.read_csv(file_path) #membaca dataset

# Tampilkan beberapa baris pertama dari dataset
print(df.head()) #menampilkan dataset pada 5 baris awal
```

```python
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.get_dummies(df) # Menghasilkan variabel dummy
df
```

```python
# Memilih fitur-fitur yang akan digunakan dalam model
predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']

# Memisahkan fitur-fitur yang dipilih dari dataframe
x = df[predictions]

# Memilih target atau label yang ingin diprediksi
y = df['Purchased']
```

```python
# Membuat objek StandardScaler untuk melakukan standarisasi data
scaler = StandardScaler()

# Menggunakan fit() untuk menghitung mean dan standard deviation dari setiap fitur di dataset
scaler.fit(x)

# Menggunakan transform() untuk mengubah data menjadi skala standar
scaled_data = scaler.transform(x)

# Membuat DataFrame baru dari data yang telah diubah, dengan kolom yang memiliki nama yang sama dengan x
scaled_data = pd.DataFrame(scaled_data, columns=x.columns)

# Menampilkan lima baris pertama dari data yang telah diubah untuk memeriksa hasil standarisasi
scaled_data.head()
```

```python
# Menggunakan train_test_split untuk membagi data menjadi set pelatihan dan pengujian
# scaled_data adalah fitur yang telah diubah skala
# y adalah target atau label yang ingin diprediksi

# test_size adalah proporsi data yang akan digunakan untuk pengujian, dalam hal ini 20%
# random_state digunakan untuk mengontrol keacakan dalam pembagian data, sehingga hasilnya dapat direproduksi
x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
# Membuat objek model logistic regression menggunakan library linear_model
model = linear_model.LogisticRegression()

# Melatih model logistic regression menggunakan data pelatihan (x_train) dan labelnya (y_train)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test) # menghitung akurasi model
```
