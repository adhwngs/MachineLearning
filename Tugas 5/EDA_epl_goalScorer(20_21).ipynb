```python
import warnings
warnings.filterwarnings('ignore')
df['goals'].mean()

sns.distplot( df[df['assists'] ==  'Right' ]['goals'], hist = False)
sns.distplot( df[df['assists'] ==  'Left' ]['goals'] , hist = False)
```

```python
df['goals'] = pd.to_numeric(df['goals'], errors='coerce')
print(df['goals'].dtype)

skill_X_preferredfoot  = pd.crosstab( df['goals'], df['assists'])
sns.heatmap(skill_X_preferredfoot)
plt.show()

new = df.groupby('assists')['goals'].mean()
new.plot(kind='bar')
plt.show()
```

```python
cluster = pd.crosstab(df['goals'] ,df['games'] )
sns.clustermap(cluster)
```
